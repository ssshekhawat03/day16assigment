```python
df['parental level of education'].value_counts()
```

```python
label=list(df['parental level of education'].value_counts().keys())
label
```

```python
values=list(df['parental level of education'].value_counts().values)
values
```

```python
plt.pie(values,labels=label,wedgeprops={'ec':'white'})
plt.show()
```

```python
ls=[0,0,0,0.2,0,0]
sns.set(rc={'figure.figsize':(7,7)})
plt.pie(values,labels=label,wedgeprops={'ec':'white'},explode=ls,autopct="%1.2f%%",shadow=True)
plt.savefig("piechart")
plt.title('parental level of education')
plt.show()
```
